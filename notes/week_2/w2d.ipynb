```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
plt.style.use('ggplot')
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
# central limit theorem when samples are drawn from same distribution
N = int(10000)
M = int(500)
x = np.random.exponential(size=M*N)
print("Estimated mean of each x: %.3f" %x.mean())
print("Estimated variance of each x: %.3f" %x.var())
plt.figure()
sns.distplot(x, bins = 100, kde = False)
plt.show()

x = np.reshape(x, (M,N))
y = np.sum(x, axis=1)
print("Estimated mean of each y: %.3f" %y.mean())
print("Estimated variance of each y: %.3f" %y.var())
plt.figure()
sns.distplot(y, bins=100, kde = False)
plt.show()
```
