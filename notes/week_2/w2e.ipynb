```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tikzplotlib
%matplotlib notebook

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
# scatter points drawn from a multivariate gaussian
N = 10000
D = 2
X = np.random.randn(N, D)

sns.scatterplot(X[:,0], X[:,1])
plt.show()
```

```python
# plot the distribution of each dimension separately
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)

sns.distplot(X[:,0], kde=True, bins=20, ax=ax1)
sns.distplot(X[:,1], kde=True, bins=20, ax=ax2)

ax1.set_ylabel("probability: $P_X(x)$")
ax1.set_xlabel("$X_1$")
ax2.set_xlabel("$X_2$")
plt.show()

# and together
fig, ax = plt.subplots()

sns.distplot(X[:,0], kde=True, bins=20, ax=ax, label="$X_1$")
sns.distplot(X[:,1], kde=True, bins=20, ax=ax, label="$X_2$")
ax.set_xlabel("$X$")
ax.set_ylabel("probability: $P_{X_i}(x)$")
plt.legend()
# plt.savefig("temp", format='eps')
# tikzplotlib.save("tmp_mytikz.tex")
plt.show()
```

```python
plt.figure()
sns.jointplot(X[:,0], X[:,1], kind='kde')
plt.show()
```
