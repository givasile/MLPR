```python
from sklearn import linear_model
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
diabetes = datasets.load_diabetes()

N_train = 200
N_val = 50
N_test = 50

# Training
X_train = diabetes.data[ : N_train]
X_val = diabetes.data[N_train : N_train + N_val]
X_test = diabetes.data[N_train + N_val : N_train + N_val + N_test]

# Split the targets into training/testing sets
y_train = diabetes.target[ : N_train]
y_val = diabetes.target[N_train : N_train + N_val]
y_test = diabetes.target[N_train + N_val : N_train + N_val + N_test]
```

```python
model = linear_model.LinearRegression()

model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_val_pred = model.predict(X_val)
y_test_pred = model.predict(X_test)
```

```python
# Print Errors

# Training set
print("Training set  : Mean squared error: %.3f" % mean_squared_error(y_train, y_train_pred))

# validation set
print("Validation set: Mean squared error: %.3f" % mean_squared_error(y_val, y_val_pred))

# Test set
print("Test set      : Mean squared error: %.3f" % mean_squared_error(y_test, y_test_pred))
```

```python
# Plot
plt.figure()
plt.scatter(np.arange(len(y_test)), y_test,  color='black')
plt.plot(np.arange(len(y_test)), y_test_pred, '-o', color='blue', linewidth=1)
plt.show()

plt.figure()
plt.plot(np.arange(len(y_test)), np.abs(y_test_pred - y_test), '-o', color='blue', linewidth=1)
plt.show()
```
